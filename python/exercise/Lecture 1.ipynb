### 1.3 = 는 '같다'가 아니다.

#### [1.3.1] 
프로그램에는 값이 하나로 정해진 숫자 뿐 아니라 프로그램이 실행되는 중에 값이 얼마든지 변할 수 있는 변수를 자주 이용합니다. 예를 들어 **x = 3**이라고 입력하면 파이썬에서는 3이라는 값을 x라는 변수에 할당합니다. 아래의 코드 셀을 실행해보세요.

````python
# *****[1.3.1] 아래의 코드를 실행하세요*****
a = 1
print(a)

b = 2
print(b)

c = a + b
print(c)
````

####[1.3.2] 
파이썬 뿐 아니라 많은 프로그램에서 **a = b**는 a와 b가 같다는 뜻이 아니라 b의 값을 변수 a에 할당(assign)한다는 뜻입니다. 예를 들어, 위의 코드 셀에서처럼 a와 b에 각각 1과 2를 할당한 후 **a = b**를 실행한다면 a에 할당된 값은 1에서 2로 바뀌게 됩니다. 참고로 파이썬에서 같다는 ==와 같이 등호를 두번 연속으로 사용하여 나타냅니다.

````python
# *****[1.3.2] 아래의 코드를 실행하세요*****
a = 1
print(a)

b = 2
print(b)

a = b
print(a)
print(b)
````

####[1.3.3] 
파이썬에서는 둘 이상의 변수에 값을 동시에 할당하는 것을 상당히 편하게 할 수 있습니다.
```
a, b = 1, 2
```
위의 코드는 a에는 1을, b에는 2를 할당합니다. 이를 이용하면, 두 변수의 값을 서로 바꾸는 것은 아래처럼 쉽게 할 수 있습니다.
```
a, b = b, a
```
예를 들어, C/C++과 같은 다른 언어에서는 위를 tem = a; a = b; b = tem;으로 해야 해서 파이썬보다 불편할 뿐 아니라 직관적이지도 않습니다.

````python
# [1.3.3] 위의 예시 코드를 참고하여 변수 company_a와 company_b에 각각 "SpaceX"과 "Tesla"를 할당하고, 두 변수에 할당된 값을 서로 바꿔주는 코드를 작성하세요. 
# print() 함수를 통해 바꿔주기 전과 후 두 변수에 할당된 값을 비교하세요.
# ***** 아래에 코드를 작성하세요*****
company_a, company_b = "SpaceX", "Tesla"
print(company_a, company_b)

company_a, company_b = company_b, company_a
print(company_a, company_b) 

# ***********************************
````

### 1.4 파이썬의 자료형

####[1.4.1] 
우리는 앞서 변수에 숫자와 문자열을 할당해보았습니다. 이처럼 변수에 할당된 값의 종류를 자료형(data type)이라고 합니다. 각 변수에 어떤 자료형의 값이 들어가 있는지에 따라 그 값을 저장하고 계산하는 방법이 다르기 때문에 이러한 값의 종류를 이해하는 것이 매우 중요합니다. 파이썬에는 다양한 자료형이 있지만 우선 숫자를 나타내는 int와 float, 문자열을 나타내는 string, 그리고 파이썬의 대표적인 기본 컨테이너인 list에 대해서 배우도록 하겠습니다. 

우선 숫자를 나타내는 자료형으로 int와 float이 있습니다. int는 integer의 약자로 정수 5, 0, -1과 같은 정수를 나타냅니다. float은 소숫점을 포함한 숫자로 5.0, 0.31, -6.1 등이 있습니다. 3과 3.0이 완전히 다른 자료형으로 구분됩니다. int를 float으로 바꿀 때는 **float()**, float을 int로 바꿀 때는 **int()**를 사용할 수 있습니다. 아래의 예시에서 **type()**이라는 명령어를 통해 각 변수에 할당된 자료형을 출력하였습니다.

````python
# *****[1.4.1] 아래의 코드를 실행하세요*****
a = 3
print(type(a))

b = 3.0
print(type(b))

# python: int와 float를 곱하면 float로 변환된다 -> AI Code에서 주의해야 할 부분!
c = a * b
print(type(c))

c = int(c)
print(type(c))

c = float(c)
print(type(c))
````

####[1.4.4] 
list는 파이썬의 대표적인 기본 컨테이너로 다양한 자료형을 유연하게 순서대로 담을 수 있는 특징이 있습니다. 예를 들어 다음과 같이 리스트를 만들 수 있습니다.

###1.5 리스트의 활용

####[1.5.1] 
리스트를 활용하기 위한 여러 명령어가 존재합니다. 우선 리스트에 새로운 요소를 추가하려면 append를 활용하면 됩니다. 예를 들어 L = [10, 20, 30, 40]에 이어서 50을 추가하려면 L.append(50)와 같이 명령하면 됩니다. 특정 요소를 지우는 것은 remove로 L.remove(30)을 하면 세 번째 요소 30이 삭제되고 각 요소가 한 칸씩 앞으로 오게 됩니다. 삭제하고자 하는 요소가 여러 번 나타나면 첫 번째 것을 삭제합니다.

````python
# *****[1.5.1] 아래의 코드를 실행하세요*****
L = [10, 20, 30, 40]
# append: 데이터를 맨 끝에 넣어준다
L.append(50)
print(L)

# remove: 인덱스가 아닌 데이터 자체를 입력한다
L.remove(30)
print(L)

L.append(20)
print(L)
L.remove(20)
print(L)
````

####[1.5.2] 
그 밖에 리스트의 요소들을 순서대로 정렬하는 sort, 순서를 뒤집는 reverse, 특정 요소의 인덱스를 알려주는 index, 특정 인덱스에 해당 요소를 입력하는 insert도 있습니다.

````python
# *****[1.5.2] 아래의 코드를 실행하세요*****
L = [1, 6, 9, 3, 5, 2]
L.sort()
print(L)

# 맨 앞 또는 맨 뒤로 element를 일관되게 추가하는 경우가 많기 때문에, elements를 통째로 뒤집어주는 reverse 많이 사용
L.reverse()
print(L)

L = [10, 20, 40, 50]
print(L.index(20))

# insert: 2개의 인자가 필요하다. 
L.insert(2, 30)
print(L)
````

````python
# string도 오름차순 정렬된다
L = ["k", "d", "f", "z", "x", "a"]
L.sort()
print(L)
````

### 1.7 if문

#### [1.7.1]
bool자료형은 True와 False 두 가지로 구성이 되며 주로 조건문이 참인지 거짓인지를 나타냅니다. 예를 들어 x == y라는 조건문이 주어졌을 때 변수 x와 y에 할당된 값이 같다면 True, 다르다면 False 값을 지니게 됩니다. 조건문을 구성하는 비교연산자에는 ==, !=, >, < 등이 있고, 논리연산자에는 and, or 등이 있습니다.

````python
type(True)
````

````python
print(2==2)
print(type(2==2))
````

````python
# *****[1.7.1] 아래의 코드를 실행하세요*****
a = 1
b = 1
c = 2

print(a == b)
print(a != b)
print(a < c)
print((a == b) and (a > c)) # 논리 연산자
print((a != c) or (a > c))
````

####[1.7.2]
if문은 함께 나오는 조건문이 True인지 False인지에 따라 코드의 실행여부를 결정합니다. else를 함께 활용하여 True인 경우에는 if문을 False인 경우에는 else문을 따르도록 프로그래밍할 수 있습니다.

if문은 for문과 함께 사용하여 for loop을 도는 동안 조건을 만족할 경우에만 특정 작업을 수행하게 하거나, 조건을 달성하면 break를 통해 for문의 loop를 끝내게 하는 데 자주 사용됩니다.

````python
multiples_3 = []
for i in range(100):
    # 1. epoch 수가 클수록 모든 학습 결과를 출력하기엔 양이 방대하기 때문에, 1000의 배수만 출력하게 함
    # 2. 보통 첫번째 학습 결과는 출력하지 않는 경우가 많아 i > 0 조건 추가
    if (i % 3 == 0) and i > 0:
        multiples_3.append(i)
print(multiples_3)
````

````python
# *****[1.8.4] 아래의 코드를 실행하세요*****
value_list = []
for row in data:
    # csv에 저장된 것은 모두 문자이므로 float type으로 변환
    value = float(row[1])
    value_list.append(value)
````

### Q4. 어떤 산업에서의 평균 밸류에이션 가장 높을까?


````python
# *****[1.8.10] 아래의 코드를 실행하세요*****
industry_list = []
for row in data:
    industry = row[3]
    if industry != '': # industry column이 공란인 기업은 제외
        if industry not in industry_list: # 겹치지 않게 industry_list에 분야를 저장
            industry_list.append(industry)
````

````python
print(industry_list)
````

````python
# *****[1.8.11] 아래의 코드를 실행하세요*****
value_avg_list = []
for industry in industry_list:
    values = []
    for row in data:
        if row[3] == industry:
            values.append(float(row[1]))
    value_avg = np.mean(values)
    value_avg_list.append(value_avg)
````

````python
# *****[1.8.12] 아래의 코드를 실행하세요*****
print(value_avg_list)
````

````python
# *****[1.8.13] 아래의 코드를 실행하세요*****
import matplotlib.pyplot as plt
plt.hist(value_avg_list, bins=20)
plt.show()
````

````python
# *****[1.8.14] 아래의 코드를 실행하세요*****
industry_max_value = industry_list[value_avg_list.index(np.max(value_avg_list))]
print(industry_max_value)
````

### Q5. 어떤 산업에 스타트업 기업이 가장 많을까?

````python
# *****[1.8.15] 이상의 내용을 활용하여 산업별 valuation의 평균을 구하고 이를 value_avg_list에 저장하세요*****
# 힌트: [1.8.11]과 동일하게 이중 for문과 if문을 활용하여 if조건문을 만족하는 경우를 세면 됩니다.

count_list = []
for industry in industry_list:
# ***** 아래에 코드를 작성하세요*****
    count = 0
    for row in data:
        if row[3] == industry:
            count += 1
    count_list.append(count)
# ***********************************
````

````python
# *****[1.8.16] 아래의 코드를 실행하세요*****
print(count_list)
````

````python
# *****[1.8.17] 아래의 코드를 실행하세요*****
import matplotlib.pyplot as plt
plt.hist(count_list, bins=20)
plt.show()
````

````python
# *****[1.8.18] 아래의 코드를 실행하세요*****
industry_max_count = industry_list[count_list.index(np.max(count_list))]
print(industry_max_count)
````
